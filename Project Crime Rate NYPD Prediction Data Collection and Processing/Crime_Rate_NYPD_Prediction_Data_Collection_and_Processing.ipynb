```python
def find_outliers_IQR(Dropped_model_data):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = Dropped_model_data[((Dropped_model_data<(q1-1.5*IQR)) | (Dropped_model_data>(q3+1.5*IQR)))]

   return outliers
```

```python
Dropped_model_data.sort_index(inplace=True)
```

```python
#First run fare_amount through the function to return a series of the outliers.

outliers = find_outliers_IQR(Dropped_model_data[["JURISDICTION_CODE"]])

print("number of outliers: "+ str(len(outliers)))

print("max outlier value: "+ str(outliers.max()))

print("min outlier value: "+ str(outliers.min()))

#outliers
```
